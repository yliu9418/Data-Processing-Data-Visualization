```python
# 5)	replace missing Values and Outliers with mean of the each variable

import pandas as pd 
import numpy as np
from matplotlib import pyplot

file = pd.read_csv('heart_risk.csv')

for i in range(len(file.columns)):
    val = file[file.columns[i]].replace(99999, np.nan).dropna(axis=0).mean()
    file[file.columns[i]] = file[file.columns[i]].replace(99999, val)
    
dataset = file
print(dataset)

#Show in Visualization after you fix missing Values and Outliers 
dataset.plot(kind='box', subplots=True, layout=(4,4), sharex=False) 
pyplot.rcParams["figure.figsize"] = 14,14
pyplot.show()
```

```python
# Compare Algorithms
from pandas import read_csv
from matplotlib import pyplot
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
import numpy
# load dataset
data = read_csv('heart_risk.csv', header=None)
df = data.drop(0)
df [:] = df [:].replace(99999, numpy.NaN)
#print(da.head(10))
array = df.values
X = array[:,0:13]
Y = array[:,13]
# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = KFold(n_splits=10, random_state=7)
    cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring) 
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std()) 
    print(msg)
    # boxplot algorithm comparison
fig = pyplot.figure() 
fig.suptitle('Algorithm Comparison') 
ax = fig.add_subplot(111) 
pyplot.boxplot(results) 
ax.set_xticklabels(names) 
pyplot.show()
```

```python
# Q-7 Standardize data (0 mean, 1 stdev)
from sklearn.preprocessing import StandardScaler 
from pandas import read_csv
from numpy import set_printoptions
data = read_csv('heart_risk.csv', header=None)
df = data.drop(0)
df [:] = df [:].replace(99999, numpy.NaN)
#print(da.head(10))
array = df.values
# separate array into input and output components
X = array[:,0:13]
Y = array[:,13]
scaler = StandardScaler().fit(X[:,1:3])
rescaledX = scaler.transform(X[:,1:3]) # summarize transformed data 
set_printoptions(precision=3) 

kfold = KFold(n_splits=10, random_state=7)
model = SVC()
results = cross_val_score(model,  np.c_[rescaledX, X[:,0:1], X[:,3:]], np.r_[Y[1:3],Y[0:1],Y[3:]], cv=kfold)
print(results.mean())
```

```python
# Q 8 --3 least contributing features using RFE test 
# Feature Extraction with RFE
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# load data
filename = 'heart_risk.csv'
names = ['Age','Sex','CP','Trestbps','Chol','Fbs','Restecg','Thalach','exang','old','peak','clope','ca','thal','Risk'] 
data = read_csv(filename, names = names)
df = data.drop(0)

array = df.values
X = array[:,0:13]
Y = array[:,13]
# feature extraction
model = LogisticRegression()
rfe = RFE(model, 3)
fit = rfe.fit(X, Y)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```
